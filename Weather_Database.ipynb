```python
import json
import pandas as pd
import random
import urllib.request
import time

from citipy import citipy
from config import weather_api_key
api_key = weather_api_key
```

```python
# 3. Create a new set of 2,000 random latitudes and longitudes.
# Let's do 4,000 for redundancy since the server seems not to know
# about as many cities as citipy does.
random_latlng = []
for i in range(4000):
    # Latitudes range from -90 to +90
    lat = -90 + random.random() * 180
    # Longitudes range from -180 to +180
    lng = -180 + random.random() * 360
    random_latlng.append([lat, lng])
```

```python
# 4. Get the nearest city using the citipy module.
cities = []
for latlng in random_latlng:
    cities.append(citipy.nearest_city(latlng[0], latlng[1]))
clean_city_names = []
countries = []
lats = []
lngs = []
max_temp = []
humidity = []
cloudiness = []
wind_speed = []
weather_desc = []
```

```python
# 5. Perform an API call with the OpenWeatherMap. 6. Retrieve the following information.
baseurl = "https://api.openweathermap.org/data/2.5/weather?q="

count = 0
count_to_get = 2000
for city in cities:
    # Don't overload the server or risk getting banned for too many requests per second.
    time.sleep(1.2)
    print(str(count) + ": " + city.city_name + " ", end="", flush=True)
    try:
        with urllib.request.urlopen(
            baseurl + city.city_name.replace(" ", "+") + f"&appid={api_key}") as url:
            data = json.loads(url.read().decode())
            lats.append(data["coord"]["lat"])
            lngs.append(data["coord"]["lon"])
            clean_city_names.append(data["name"])
            countries.append(data["sys"]["country"])
            max_temp.append(data["main"]["temp_max"])
            humidity.append(data["main"]["humidity"])
            cloudiness.append(data["clouds"]["all"])
            wind_speed.append(data["wind"]["speed"])
            weather_desc.append(data["weather"][0]["description"])
            count += 1
            if count >= count_to_get:
                break
    except urllib.error.HTTPError:
        # Maybe the server doesn't know about this city, let's skip it.
        pass
    except urllib.error.URLError as e:
        # Network failure, let's skip this one.
        pass
```

```python
# 7. Add the data to a new DataFrame.
df = pd.DataFrame({
    "City": clean_city_names,
    "Country": countries,
    "Lat": lats,
    "Lng": lngs,
    "Max Temp": max_temp,
    "Humidity": humidity,
    "Cloudiness": cloudiness,
    "Wind Speed": wind_speed,
    "Current Description": weather_desc
})

df.head(10)
```

```python
# Export the DataFrame as a CSV file.
df.to_csv("WeatherPy_Database.csv")
```
